```r
library(ggplot2)
library(dplyr)
```

#### CG 甲基化

```r
gf=gzfile("./results/test.CG.txt.gz",'rt')
intersectedData=read.table(
    gf,
    header=F,sep="\t"
)
plotData=intersectedData %>% mutate(
        #* inplace unmatch window with 0
        V10=ifelse(V9==".",0,as.character(V9)),
        V11=as.numeric(V10)
    )  %>% group_by(
        V1,V2,V3,V5
    ) %>% summarise(meanval=mean(V11))

p=plotData %>%
    group_by(
        V5
    ) %>%
    summarise(windowVal = mean(meanval)) %>%
    ggplot(
        aes(x = V5, y = windowVal)
    ) +
    geom_line(
        color='#008B81',
        size=2
    )+
    theme_classic()+
    theme(
        panel.background = element_blank(),
        panel.grid = element_blank(),
        axis.title = element_text(size=16),
        axis.text = element_text(size=12),
    )+
    geom_vline(
        xintercept =26,
        linetype=2
    )+geom_vline(
        xintercept =41,
        linetype=2
    )+
    xlab("Gene body")+ylab("mCG ratio")
p    
```

```r
png(filename = "results/test_CG.png",width = 400,height = 400)
print(p)
dev.off()
```
